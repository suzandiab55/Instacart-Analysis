# Exercise 2.1

## Importing Libraries

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

## Importing File

```python
# Importing ords_prods_all file
ords_prods_all = pd.read_pickle(r'/Users/suzandiab/Documents/Instacart Basket Analysis/02 Data/Prepared Data/ords_prods_all.pkl')
```

# Exercise 2.2 

## Creating bar chart

```python
# Creating bar chart for orders day of week column (largest to smallest)
ords_prods_all['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Creating bar chart for orders day of week column and sorting by index order
ords_prods_all['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Adding color to bar chart 
bar = ords_prods_all['orders_day_of_week'].value_counts().plot.bar(color = ['red', 'red', 'orange', 'orange', 'orange', 'yellow', 'yellow'])
plt.title('Frequency of Orders for Each Day of the Week', fontweight='bold')
plt.xlabel('Days of the Week', fontweight='bold')
plt.ylabel('Order Count', fontweight='bold')
plt.ticklabel_format(style='plain', axis='y')
```

```python
# Export bar chart as image file
bar.figure.savefig(r'/Users/suzandiab/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/bar_orders_all_dow.png', bbox_inches='tight')
```

## Creating Histogram

```python
# Creating histogram of prices column
ords_prods_all['prices'].plot.hist(bins = 25)
```

```python
# Increasing level of granularity by adding more bins, can see more variety and saving it
hist = ords_prods_all['prices'].plot.hist(bins = 70)
plt.title('Number of Products According to Price', fontweight='bold')
plt.xlabel('Price Ranges', fontweight='bold')
plt.ylabel('Product Count', fontweight='bold')
plt.ticklabel_format(style='plain', axis='y')
```

```python
# Exporting histogram as image file
hist.figure.savefig(r'/Users/suzandiab/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/hist_prices_all.png', bbox_inches='tight')
```

## Creating Scatterplot

```python
# Creating scatterplot of prices column
scatter_prices = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_all)
```

```python
# Exporting scatterplot as image file
scatter_prices.figure.savefig(r'/Users/suzandiab/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/scatter_prices_all.png', bbox_inches='tight')
```

# Creating Line Chart with subset

```python
# Creating random sample by creating list holding true/false values to the test np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_all)) <= 0.7
```

```python
# Displaying lsit
dev
```

```python
# Store 70% of sample in big df
big = ords_prods_all[dev]
```

```python
# Store 30% of sample in small df
small = ords_prods_all[~dev]
```

```python
# Checking length of df
len(ords_prods_all)
```

```python
# Adding both df's to check length is accurate
len(big) + len(small)
```

```python
# Reducing sample to only necessary columns for line chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create line chart for orders dow column and price column
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
plt.title('Price Throughout Days of the Week', fontweight='bold')
plt.xlabel('Day of the Week', fontweight='bold')
plt.ylabel('Price', fontweight='bold')
```

Most money is spent on Friday and Saturday.

```python
# Exporting line chart as image file
line.figure.savefig(r'/Users/suzandiab/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/line_chart_all.png', bbox_inches='tight')
```

## Creating scatterplot between age and spending power (income)

```python
# Creating scatterplot of age and income columns
scatter_age_income = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_all)
plt.title('Relationship between Age and Spending Power', fontweight='bold')
plt.xlabel('Age', fontweight='bold')
plt.ylabel('Income', fontweight='bold')
```

The scatterplot shows:
    1. People aged 18-40 have a lower income when compared those above 40.
    2. People aged 18-40 have an income of less than 400,000.
    3. People above the age of 40 have an income of up to 600,000.
